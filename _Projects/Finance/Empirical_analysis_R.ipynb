## Importing data from Quandl

```r
library(Quandl)
data <- Quandl('ECB/EURUSD', type = "xts")
```

```r
length(data)
```

```r
summary(data)
```

```r
plot(data)
```

```r
plot(diff(data))
```

```r
acf(data)
```

```r
pacf(data)
```
